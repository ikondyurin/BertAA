```python
import pandas as pd

lime_features = pd.read_csv("/Users/ivankondyurin/Library/CloudStorage/OneDrive-Личная/lime_features.csv")
att_features = pd.read_csv("/Users/ivankondyurin/Library/CloudStorage/OneDrive-Личная/att_features.csv")
```

```python
feat = list(lime_features['0'])

from collections import Counter

feat_count = Counter(feat)
feat_count = feat_count.most_common()
#feat_count = dict(feat_count[:25])
```

```python
feat2 = list(att_features['0'])

from collections import Counter

feat2_count = Counter(feat2)
feat2_count = feat2_count.most_common()
feat2_count = dict(feat2_count[:25])
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 10))
plt.bar(feat_count.keys(), feat_count.values(), width=0.5, orientation=u'vertical',)

plt.xlabel("LIME features")
plt.ylabel("Counts in 100 samples")
plt.title("25 most common LIME features selected from 100 samples")
plt.xticks(rotation=45)
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 10))
plt.bar(feat2_count.keys(), feat2_count.values(), width=0.5, orientation=u'vertical',)

plt.xlabel("Attention-based features")
plt.ylabel("Counts in 100 samples")
plt.title("25 most common attention-based features selected from 100 samples")
plt.xticks(rotation=45)
plt.show()
```
