```python
from collections import Counter
layers = dict()
allnames = []

for index, i in enumerate(atan["Important Tokens"]):
    layer = str(int(atan.iloc[[index]]["Layer"]))
    types = i.lower().split(",")
    types = [x.strip() for x in types]
    #print(layer, types)
    allnames += types
    if layer in layers:
        layers[layer] += types
    else:
        layers[layer] = types

for key in layers.keys():
    layers[key] = Counter(layers[key])
    #print(layers[key])

#print(layers)

print(Counter(allnames))
    
```
